```python
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------
# Replace these with your actual CSV file paths
# ----------------------------------------------------
male_train_path = "/scratch/sshuvo13/project_shared_folder_bspml_1/segments_30s/features/train_test_separated_and_combined/male/feature33/train_data.csv"
male_test_path  = "/scratch/sshuvo13/project_shared_folder_bspml_1/segments_30s/features/train_test_separated_and_combined/male/feature33/test_data.csv"
female_train_path = "/scratch/sshuvo13/project_shared_folder_bspml_1/segments_30s/features/train_test_separated_and_combined/female/feature33/train_data.csv"
female_test_path  = "/scratch/sshuvo13/project_shared_folder_bspml_1/segments_30s/features/train_test_separated_and_combined/female/feature33/test_data.csv"
# ----------------------------------------------------

# Load datasets
male_train = pd.read_csv(male_train_path)
male_test  = pd.read_csv(male_test_path)
female_train = pd.read_csv(female_train_path)
female_test  = pd.read_csv(female_test_path)

# The label column is assumed to be the last column
label_col = male_train.columns[-1]

datasets = {
    "Male Train": male_train,
    "Male Test": male_test,
    "Female Train": female_train,
    "Female Test": female_test
}

# Plot histograms
plt.figure(figsize=(12, 8))

for i, (name, df) in enumerate(datasets.items(), 1):
    plt.subplot(2, 2, i)
    df[label_col].value_counts().plot(kind='bar')
    plt.title(f"{name}  Normal vs Apnea")
    plt.xlabel("Class")
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```
