```python
from xml.etree import ElementTree as ET
import pandas as pd
import os
import re
```

```python
rml_dir = "/scratch/sshuvo13/all_rml_files/rml_files/"
```

```python
tree = ET.parse("./00001010-100507.rml")
```

```python
all_respiratory_events = []
for resp_event in tree.findall(".//{http://www.respironics.com/PatientStudy.xsd}Event[@Family='Respiratory']"):
    single_event_attributes = {}
    single_event_attributes['start_time'] = resp_event.attrib['Start']
    single_event_attributes['duration'] = resp_event.attrib['Duration']
    single_event_attributes['event_name'] = resp_event.attrib['Type']
    all_respiratory_events.append(single_event_attributes)
```

```python
output_dir = './new_apnea_events_identified_in_csv'
os.makedirs(output_dir, exist_ok = True)


for filename in os.listdir(rml_dir):
    if filename.endswith('.rml'):
        # rml_file_paths.append(os.path.join(data_dir, filename))
        # rml_file_names.append(filename)
        rml_file_path = os.path.join(rml_dir, filename)
        tree = ET.parse(rml_file_path)
        all_respiratory_events = []
        for resp_event in tree.findall(".//{http://www.respironics.com/PatientStudy.xsd}Event[@Family='Respiratory']"):
            single_event_attributes = {}
            single_event_attributes['start_time'] = resp_event.attrib['Start']
            single_event_attributes['duration'] = resp_event.attrib['Duration']
            single_event_attributes['event_name'] = resp_event.attrib['Type']
            all_respiratory_events.append(single_event_attributes)

        match =  re.search(r'([0-9]{8}-[0-9]{6})(?=\.rml)', filename)
        patient_id = match.group(1)
        
        dff = pd.DataFrame(all_respiratory_events)
        dff.to_csv(os.path.join(output_dir, f"{patient_id}.csv"), index = False)
    
```
